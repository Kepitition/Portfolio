```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("nyc_taxi.csv")


def cusum_algorithm(data, threshold= 15000):
    n = len(data)
    avg = data.mean()
    cusum = [0] * n
    change_points = []

    for i in range(1, n):
        cusum[i] = max(0, (cusum[i-1] + data[i]) - avg - threshold)
        if cusum[i] > 0:         
            change_points.append(i)

    return change_points


change_points = cusum_algorithm(data['value'])

plt.figure(figsize=(12, 3))
plt.plot(data['value'], label='Value')
plt.plot(change_points, data['value'][change_points], 'ro', markersize=4, label='Change Point')
plt.xlabel('Zaman')
plt.ylabel('Passenger Value')
plt.title('Change Point Detection Test Results With Cusum Algorithm Treshold = 15000')
plt.legend()
plt.show()
```

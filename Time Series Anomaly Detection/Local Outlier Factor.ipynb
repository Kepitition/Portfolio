```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor


data = pd.read_csv("nyc_taxi.csv")

lof = LocalOutlierFactor(n_neighbors=3000, contamination=0.001)


anomaly_scores = lof.fit_predict(data[['value']])


data['anomaly_score'] = anomaly_scores


plt.figure(figsize=(12, 3))
plt.plot(data['value'], label='Value')
plt.plot(data[data['anomaly_score'] == -1]['value'], 'ro', markersize=4, label='Anomaly')
plt.xlabel('Timestamp')
plt.ylabel('Passenger Value')
plt.title('Local Outlier Factor Without Data Scale')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import OneClassSVM


data = pd.read_csv("nyc_taxi.csv")


nu = 0.95
one_class_svm = OneClassSVM(kernel='rbf', nu=nu)
one_class_svm.fit(data[['value']])

anomaly_scores = one_class_svm.decision_function(data[['value']])

anomalies = data[anomaly_scores < -3000]

plt.figure(figsize=(12, 6))
plt.plot(data['value'], label='Passenger')
plt.plot(anomalies.index, anomalies['value'], 'ro', markersize=4, label='Passenger Anomaly')
plt.xlabel('Timestamp')
plt.ylabel('Passenger Value')
plt.title('One Class SVM Test')
plt.legend()
plt.show()
```

```python
for i in anomaly_scores:
    if i < -9000:
        print(i)
```
